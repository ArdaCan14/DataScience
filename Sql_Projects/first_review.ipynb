```python
%%sql
SELECT AVG(games.HOME_TEAM_WINS) AS Game_Wins,teams.NICKNAME FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Game_Wins DESC;
```

```python
# When we examine the code above cell,we are able to see Jazz that won most games at home AVARAGELY! till 2019
```

```python
%%sql
SELECT COUNT(games.HOME_TEAM_WINS) AS Game_Wins,teams.NICKNAME,teams.YEARFOUNDED FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Game_Wins DESC;
```

```python
# When we take a look above cell The team that won most times is Miami Heat. 
```

```python
#This queries show us the team that has best avarage winnig rate is not same with the the team has most winnig games.
```

```python
%%sql
SELECT AVG(NOT HOME_TEAM_WINS) AS AVG_Visitor_Wins,teams.NICKNAME FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID 
GROUP BY teams.NICKNAME ORDER BY AVG_Visitor_Wins DESC;
```

```python
# As we are able to see above cell,Most successful visitor team is Phoenix Suns,considiring avarage.
```

```python
%%sql
SELECT COUNT(NOT HOME_TEAM_WINS) AS Visitor_Wins,teams.NICKNAME,teams.MIN_YEAR,teams.MAX_YEAR FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Visitor_Wins DESC;
```

```python
#As it can be seen considering the count of visitor team wins the most successful team is Miami Heat
```
