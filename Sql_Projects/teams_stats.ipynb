```python
%%sql
SELECT teams.NICKNAME,AVG(games.PTS_home)AS Home_PTS FROM games lEFT OUTER JOIN games_details  ON games.GAME_ID = games_details.GAME_ID 
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Home_PTS DESC;
```

```python
# As it can be seen the team that has highest amount of point in its own arena is Golden State Warrious
```

```python
%%sql
SELECT teams.NICKNAME,AVG(games.PTS_away) AS Visitor_PTS FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Visitor_PTS DESC;
```

```python
#As it can be seen in the above querry the team that has highest amount of away point is Los Angeles Lakers.
```

```python
%%sql
SELECT AVG(games.PTS_home + games.PTS_away) AS total_pts,teams.NICKNAME FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY total_pts DESC; 
```

```python
#As it can be seen above querry the theam that has highest amounr of point as visitor or home team is Golden State Warrious
```

```python
%%sql
SELECT teams.NICKNAME,AVG(FG3_PCT_home + FG3_PCT_away) AS AVG_3_POINT FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY AVG_3_POINT DESC; 
```

```python
#Golden State Warrious has the highest 3pt rate between 2003-2019
```

```python
%%sql
SELECT AVG(games.AST_home + games.AST_away) AS Avarage_Asist,teams.NICKNAME FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Avarage_Asist DESC; 
```

```python
#Golden State Warrious has the highest asist rate between 2003-2019
```

```python
%%sql
SELECT AVG(games.REB_home + games.REB_away) AS Total_Rebound,teams.NICKNAME FROM games LEFT OUTER JOIN games_details ON games.GAME_ID = games_details.GAME_ID
LEFT OUTER JOIN teams ON games_details.TEAM_ID = teams.TEAM_ID GROUP BY teams.NICKNAME ORDER BY Total_Rebound DESC; 
```
