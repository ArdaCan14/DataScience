```python
%%sql
SELECT Cus.Country,AVG(Price) Avarage FROM Products Pro LEFT OUTER JOIN OrderDetails Orde ON Pro.ProductID = Orde.ProductID
LEFT OUTER JOIN Orders Ord ON Orde.OrderID = Ord.OrderID
LEFT OUTER JOIN Customers Cus ON Ord.CustomerID = Cus.CustomerID WHERE Cus.Country = 'USA'; 
```

```python
%%sql
SELECT AVG(Price),CategoryName FROM Products LEFT OUTER JOIN Categories ON Products.CategoryID = Categories.CategoryID GROUP BY CategoryName;
```

```python
%%sql
SELECT SUM(p.Price) Total_Sales,emp.FirstName FROM Products p LEFT OUTER JOIN OrderDetails orde ON p.ProductID = orde.ProductID
LEFT OUTER JOIN Orders ord ON orde.OrderID = ord.OrderID 
LEFT OUTER JOIN Employees emp ON ord.EmployeeID = emp.EmployeeID GROUP BY emp.EmployeeID ORDER BY Total_Sales DESC;
```

```python
%%sql
SELECT Cus.Country,Cat.CategoryName,AVG(Pro.Price) FROM Orders Ord LEFT OUTER JOIN Customers Cus ON Ord.CustomerID = Cus.CustomerID
LEFT OUTER JOIN OrderDetails Orde ON Ord.OrderID = Orde.OrderID
LEFT OUTER JOIN Products Pro ON Orde.ProductID = Pro.ProductID
LEFT OUTER JOIN Categories Cat ON Pro.CategoryID = Cat.CategoryID WHERE Country = 'Germany' GROUP BY CategoryName;
```

```python
%%sql
SELECT Cat.CategoryName,AVG(Pro.Price) FROM Orders Ord LEFT OUTER JOIN Customers Cus ON Ord.CustomerID = Cus.CustomerID
LEFT OUTER JOIN OrderDetails Orde ON Ord.OrderID = Orde.OrderID
LEFT OUTER JOIN Products Pro ON Orde.ProductID = Pro.ProductID
LEFT OUTER JOIN Categories Cat ON Pro.CategoryID = Cat.CategoryID WHERE Country = 'Germany' OR Country = 'USA'  GROUP BY CategoryName;
```

```python
%%sql
SELECT Cus.CustomerName,MAX(Pro.Price) Maximum FROM Products Pro LEFT OUTER JOIN OrderDetails Orde ON Pro.ProductID = Orde.ProductID
LEFT OUTER JOIN Orders Ord ON Orde.OrderID = Ord.OrderID 
LEFT OUTER JOIN Customers Cus ON Ord.CustomerID = Cus.CustomerID WHERE Ord.OrderDate LIKE '%1997'
GROUP BY Cus.CustomerName ORDER BY Maximum DESC;
```

```python
%%sql
SELECT Emp.FirstName,COUNT(Ord.OrderID) Order_Number FROM  Orders Ord LEFT OUTER JOIN Employees Emp ON Ord.EmployeeID = Emp.EmployeeID
WHERE OrderDate LIKE '%1997%' GROUP BY FirstName ORDER BY Order_Number DESC;
```
